```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data: intervals, ranges, and values
intervals = [1, 2, 3, 4, 5,6,7,8]
ranges = [10, 20, 30, 40, 50, 60, 65, 64]
values = [15, 25, 35, 45, 55, 60, 65, 64]

# Plotting the line graph
plt.figure(figsize=(10, 6))
plt.plot(intervals, values, marker='o', linestyle='-', color='b', label='Values')

# Sample data
x = np.arange(1, 11)  # Example x values
y = np.random.randint(1, 100, size=10)  # Example y value

# Plot horizontal lines at intervals of 1 along the y-axis
for i in range(1, max(y) + 1):
    plt.hlines(i, xmin=min(x), xmax=max(x), colors='gray', linestyles='-', alpha=0.5)
# Adding range as error bars or shaded area
#plt.errorbar(intervals, values, yerr=ranges, fmt='none', ecolor='gray', alpha=0.5, capsize=5)

# Customizing the plot
plt.xlabel('Intervals')
plt.ylabel('Values')
plt.title('Line Graph with Interval Ranges and Values')
plt.legend()
plt.grid(True)
plt.xticks(intervals)  # Set x-axis ticks to match intervals
plt.tight_layout()

# Show plot
plt.show()
```
